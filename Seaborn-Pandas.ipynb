```python
#importar as bibliotecas
import seaborn as sns
import pandas as pd
```

```python
#criar o DataFrame Vendas realizadas por mês
df = pd.DataFrame({'Mês':['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez'],
                 'Resultado':[1501,1200,1790,1100,1500,1000,1000,1218,1590,1300,1900,3200]},index = list(range(1,13)))
```

```python
#Exibir o DataFrame na Horizontal em ordem decresente de Resultdos
df.sort_values('Resultado', ascending=0).T
```

```python
# Exibir DataFrame na vertical
df
```

```python
#Plotar DataFrame
grafico1 = sns.barplot(x='Resultado', y='Mês',data=df, color='#4AAE8A')
```

```python
grafico2 = sns.barplot(y='Resultado', x='Mês',data=df, color='#4AAE8A')
```

```python
grafico3 = sns.boxenplot(y='Resultado', x='Mês',data=df, color='#4AAE8A')
```

```python
grafico4 = sns.catplot(y='Resultado', x='Mês',data=df, color='#4AAE8A')
```

```python
grafico5 = sns.jointplot(y='Resultado', x='Mês',data=df, color='#4AAE8A')
```

```python
grafico6 = sns.stripplot(y='Resultado', x='Mês',data=df, color='#909090')
```

```python
grafico7 = sns.catplot(y='Resultado', x='Mês',data=df, color='blue', kind="point")
```

```python
grafico8 = sns.histplot(y='Resultado', x='Mês',data=df, color='#2C4D71')
```

```python
grafico9 = sns.relplot(y='Resultado', x='Mês',data=df, color='red')
```

```python
grafico10 = sns.lineplot(y='Resultado', x='Mês',data=df, color='#2C4D71')
```
